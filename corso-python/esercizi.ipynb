```python
numero = 173
primo = True
for divisore in range(2, numero // 2):
    if numero%divisore == 0:
        primo = False
        break
if primo:
    print(f'{numero} è un numero primo')
else:
    print(f'{numero} non è un numero primo')
```

```python
def isPrimo(numero):
    for divisore in range(2, int(numero ** 0.5) + 1):
        if numero % divisore == 0:
            return False
    return True

numeriPrimiVisualizzati = 0
NUMERI_DA_VISUALIZZARE = 100
numero = 2

while numeriPrimiVisualizzati <= NUMERI_DA_VISUALIZZARE:
    if isPrimo(numero):
        print(numero, end = ' ')
        numeriPrimiVisualizzati += 1
    numero += 1
```
