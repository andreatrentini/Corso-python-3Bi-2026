```python
def isPrimo(numero):
    for divisore in range(2, int(numero ** 0.5) + 1):
        if numero % divisore == 0:
            return False
    return True

numeriPrimiVisualizzati = 0
NUMERI_DA_VISUALIZZARE = 100
numero = 2

while numeriPrimiVisualizzati <= NUMERI_DA_VISUALIZZARE:
    if isPrimo(numero):
        print(numero, end = ' ')
        numeriPrimiVisualizzati += 1
    numero += 1
```

# Tuple
Scrivere una funzione che calcoli la distanza fra due punti: p1 e p2
p1 e p2, parametri della funzione distanza contengono le coordinate x e y

```python
import math
def distanza(p1, p2):
    x1 = p1[0]
    x2 = p2[0]
    y1 = p1[1]
    y2 = p2[1]
    x = abs(x2 - x1)
    y = abs(y2 - y1)
    return math.sqrt(x ** 2 + y ** 2)

print(distanza((2,2), (6,5.5)))
```
